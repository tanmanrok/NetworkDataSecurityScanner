### Missing Data

```python
# Find missing data
data_missing = pd.concat([data_path.isnull().sum(), 100 * data_path.isnull().mean()], axis=1)
data_missing.columns = ['Data Missing', '%']
data_missing.sort_values(by=['Data Missing', '%'], ascending=False)
```

```python
# Find missing Label
label_missing = pd.concat([label_path.isnull().sum(), 100 * label_path.isnull().mean()], axis=1)
label_missing.columns = ['Label Missing', '%']
label_missing.sort_values(by=['Label Missing', '%'], ascending=False)
```

```python
CIC_missing = pd.concat([CICData.isnull().sum(), 100 * CICData.isnull().mean()], axis=1)
CIC_missing.columns = ['CICData Missing', '%']
CIC_missing.sort_values(by=['CICData Missing', '%'], ascending=False)
```

### Change Label Number to String

```python
# using the read me file to change the label numbers to strings
label_path['Label'] = label_path['Label'].replace({0: 'Benign', 1: 'Analysis', 2: 'Backdoor', 3: 'DoS', 4: 'Exploits', 5: 'Fuzzers', 6: 'Generic', 7: 'Reconnaissance', 8: 'Shellcode', 9: 'Worms'})
label_path.head()
```

### Cleaning Training data

```python
# Cleaning Training data
# No missing data to clean
# The 0 values are needed information for the model
# Removing duplicate rows if any
CICData = CICData.drop_duplicates()
```

```python
# Removing unneeded columns from CICData
columns_to_drop = ['Flow ID', 'Src IP', 'Src Port', 'Dst IP', 'Timestamp']
columns_to_drop = [col for col in columns_to_drop if col in CICData.columns]
CICData = CICData.drop(columns=columns_to_drop)
CICData.head()
```

```python
# Save cleaned data to processed folder
output_path = '../data/processed/CICData_cleaned.csv'
os.makedirs('../data/processed', exist_ok=True)
CICData.to_csv(output_path, index=False)
print(f"Cleaned data saved to: {output_path}")
print(f"Shape: {CICData.shape}")
```

## Info about labels  

- Flow ID = The log ID from the scan
- Src IP = Senders IP
- Src Port = Senders Port used
- Dst IP = Destination IP 
- Dst Port = Destination Port
- Protocol = Communication Protocol Number (1:ICMP, 6:TCP, 17:UDP, 41:IPv6 encapsulation, 47:GRE) **More Details Below**
- Timestamp
- Flow Duration = Time of Flow or Commination in ms
- Fwd = Sent
- Bwd = Received 
- IAT = Inter-Arrival Time
- PSH = Push used in TCP 
- URG = Urgent Flag
- FIN = Finish Tag
- SYN = Haft-open Flag
- RST = Reset 
- ACK = Acknowledgement 
- CWR = Congestion Window Reduced 
- ECE = Electrical and Computer Engineering
- Init = Initial
- Win = Window
- Pkts = Packets
- Seg = Segment
